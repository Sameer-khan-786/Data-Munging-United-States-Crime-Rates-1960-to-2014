### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv). 

### Step 3. Assign it to a DataFrame variable called crime.

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"
crime = pd.read_csv(url)
crime
```

### Step 4. What is the type of the columns?

```python
print(crime.dtypes)
```

##### Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.

### Step 5. Convert the type of the column Year to datetime64

```python
crime['Year'] = pd.to_datetime(crime['Year'], format='%Y')
print(crime.dtypes)
```

### Step 6. Set the Year column as the index of the dataframe (you may want to use set_index())

```python
crime.set_index('Year', inplace=True)
# Display the DataFrame after setting 'Year' as the index
crime
```

### Step 7. Delete the Total column

```python
crime.drop(columns=['Total'], inplace=True)
# Total column is Dropped
crime
```

### Step 8. Group the year by decades and sum the values
#### Hint: 
#### To get the decade, you can integer-divide the year by 10 and then multiply by 10. You can use '//' for floor division. 
#### Pay attention to the Population column number, summing this column is a mistake

```python
crime_decades = crime.groupby((crime.index.year//10)*10).sum()
# Data grouped by decades and summed up
crime_decades
```

### Step 9. What is the most dangerous decade to live in the US?

```python
most_dangerous_decade = crime_decades.idxmax()['Violent']
print(f"The most dangerous decade in the US for violent crimes was in the {most_dangerous_decade}s.")
```

# Exercise 02: Drinks

### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv). 

### Step 3. Assign it to a DataFrame variable called drinks.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"
drinks = pd.read_csv(url)
drinks
```

### Step 4. Which continent drinks more beer on average?

```python
beer_avg = drinks.groupby('continent')['beer_servings'].mean().idxmax()
print(f"The continent that drinks the most beer on average is {beer_avg}.")
```

### Step 5. For each continent print the statistics for wine consumption.

```python
wine_stats = drinks.groupby('continent')['wine_servings'].describe()
wine_stats
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
alcohol_mean = drinks.groupby('continent').mean()
alcohol_mean
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
alcohol_median = drinks.groupby('continent').median()
alcohol_median
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
spirit_stats = drinks.groupby('continent')['spirit_servings'].agg(['mean', 'min', 'max']).reset_index()
spirit_stats
```
